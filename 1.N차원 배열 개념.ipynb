Section 1. N차원 배열 생성

```python
import numpy as np
```

```python
# 1차원 배열
arr = np.array([1,2,3])
print(arr)

# 2차원 배열
arr1 = np.array([[1,2,3], [4,5,6]])
print(arr1)
```

```python
type([1,2,3])
type(arr1)

# arr1에 대하여 ndarray라고 나오는 것을 볼 수 있다. 다차원 행렬임을 숙지할 것.
```

```python
tpl = (4,5,6)
arr = np.array(tpl)
print(arr)

lst = [1,2,3]
arr = np.array(lst)
print(arr)

lst2 = [[1,2,3], [4,5,6]]
arr2 = np.array(lst2)
print(arr2)
```

```python
# Shape
arr1 = np.array([1,2,3])
arr2 = np.array([[1,2,3],[4,5,6]])

print(arr1.shape, arr2.shape) # 원소가 3개인 1차원 배열, 행이 2개이고, 열이 3개인 행렬 
```

```python
# ndim
print(arr.ndim, arr2.ndim) # 차원에 대한 함수... 1차원 , 2차원 출력 
```

```python
# size
print(arr1.size, arr2.size) # 원소의 개수.. 원소가 3개, 6개
```
